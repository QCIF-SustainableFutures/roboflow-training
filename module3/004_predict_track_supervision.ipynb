## Deploy a Model

This notebook will cover the basics of deploying a model. We will be using a new package by Roboflow called 'supervision', which is full of utilities to work with computer vision datasets and projects.

```python
# Install the required packages
# Open a terminal and run the following command
# pip install supervision inference
```

### Import Supervision, Open CV and Inference


```python
import supervision as sv
import cv2
from inference import get_model
```

### Read a sample image located in our directory

```python
#image = cv2.imread("data-004/2021-best-cars-ford-mustang-hero-desktop.jpg")
image = cv2.imread("data-004/red_kangaroo-24c18ab08dc145f1a798abd4b820390a.jpg")
```

### Infer / Predict using a base model


```python
model = get_model(model_id="yolov8n-640")
results = model.infer(image)[0]
detections = sv.Detections.from_inference(results)
```

### Annotate the image using the bounding boxes and plot the image

```python
annotated_image = sv.BoxAnnotator().annotate(
    scene=image.copy(), detections=detections
)
annotated_image = sv.LabelAnnotator().annotate(
    scene=annotated_image, detections=detections
)
sv.plot_image(annotated_image)
```
